```python
import pandas as pd

from tensortrade.features.scalers import MinMaxNormalizer, ComparisonNormalizer, PercentChangeNormalizer
from tensortrade.features.stationarity import FractionalDifference
from tensortrade.features.indicators import TAlibIndicator

ohlcv_data = pd.read_csv('./data/Coinbase_BTCUSD_1h.csv', skiprows=1)
ohlcv_data = ohlcv_data[['open','high','low','close','volume']]

# ohlcv_data
```

```python
taliber = TAlibIndicator(indicators=["EMA", "RSI", "BBANDS"])
```

```python
taliber.transform(ohlcv_data)
```

```python
normalizer = PercentChangeNormalizer()
obs = normalizer.transform(ohlcv_data)

difference = FractionalDifference(difference_order=0.5, inplace=True)
difference.reset()
difference.transform(obs)
```
